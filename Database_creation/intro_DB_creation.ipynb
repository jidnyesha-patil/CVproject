```python
import cv2
import time
import os
import pandas as pd
```

```python
# Creating a directory to store names and 
# number of pictures for each kid
def read_list():
    doc1 = {"Name":[] , "Count":[]}
    path = 'C:\\Users\\rushi\\Documents\\CV_project\\Intro_DB\\'
    name_list = os.listdir(path)
    for name in name_list:
        doc1["Name"].append(name)
        new_path = path + name + '\\'
        n = os.listdir(new_path)
        doc1["Count"].append(len(n))
    l = []
    for k,v in doc1.items():
        l.append(doc1[k])
    return l
```

```python
def save_image(name,path='C:\\Users\\rushi\\Documents\\CV_project\\Intro_DB\\'):   
    vid = cv2.VideoCapture(0)
    if (vid.isOpened() == False):
        print("Error reading video file")

    # We need to set resolutions.
    # so, convert them from float to integer.
    frame_width = int(vid.get(3))
    frame_height = int(vid.get(4))

    size = (frame_width, frame_height)
    path = path + name
    #print(path)

    if not os.path.exists(path):
        os.makedirs(path)
    path = path + "\\"
    
    
    t_start = time.time()
    t_end = time.time()
    t_elapsed = t_end - t_start
    while(t_elapsed != 30 ):

        # Capture the video frame
        # by frame
        ret, frame = vid.read()
        f = input("Continue?[y/n]")
        if f == 'y' or f == 'Y':
            if ret == True:

                # Display the frame
                # saved in the file
                check_name = read_list()
                print("check_name:", check_name)
    
                # For adding images to pre-existing folders
                # The Count variable will get updated for these folders
                # To ensure old and new data for each kid gets saved together in one folder
                name2 = 0
                if name in check_name[0]:
                    i = check_name[0].index(name)
                    name2 = check_name[1][i] + 1
                    #print("old_count =", check_name[1][i])
                else:
                    name2 +=1
                #print("name2 = ", name2) 
                # Path for saving the database
                temp = path+name+str(name2)+".jpg"
                result = cv2.imwrite(temp, frame)
            
        elif f == "n" or f == "N":
            break
       
        
        # Display the resulting frame
        # the 'q' button is set as the
        # quitting button you may use any
        # desired button of your choice
        #if cv2.waitKey(1) & 0xFF == ord('q'):
            #break
     
        t_end = time.time()
        t_elapsed = t_end - t_start
    
    # After the loop release the cap object
    vid.release()
    #result.release()
    # Destroy all the windows
    cv2.destroyAllWindows()
```

```python
f1 = input("Continue?(y/n)")
while f1 == "y" or f1 == "Y":
    name = input("Enter the name of the kid:")
    save_image(name)
    f1 = input("Continue?(y/n)")
```
